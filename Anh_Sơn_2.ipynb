$$
Trong đó:
+ FV: Future value

# **ichimoku cloud**

````python
import pandas as pd
import matplotlib.pyplot as plt

def ichimoku(df, n1=9, n2=26, n3=52):
    # Calculate components
    df['tenkan_sen'] = (df['High'].rolling(n1).max() + df['Low'].rolling(n1).min()) / 2
    df['kijun_sen'] = (df['High'].rolling(n2).max() + df['Low'].rolling(n2).min()) / 2
    df['senkou_span_a'] = ((df['tenkan_sen'] + df['kijun_sen']) / 2).shift(n2)
    df['senkou_span_b'] = ((df['High'].rolling(n3).max() + df['Low'].rolling(n3).min()) / 2).shift(n2)
    df['chikou_span'] = df['Close'].shift(-n2)  # Shifted backwards

    return df

# Example Usage:
df = pd.read_csv('AAPL.csv')
df = ichimoku(df)

# Plotting
plt.figure(figsize=(20, 6))
plt.plot(df['tenkan_sen'], label='Tenkan-sen')
plt.plot(df['kijun_sen'], label='Kijun-sen')
plt.fill_between(df.index, df['senkou_span_a'], df['senkou_span_b'], where=df['senkou_span_a'] >= df['senkou_span_b'], color='green', alpha=0.2)
plt.fill_between(df.index, df['senkou_span_a'], df['senkou_span_b'], where=df['senkou_span_a'] < df['senkou_span_b'], color='red', alpha=0.2)
plt.plot(df['chikou_span'], label='Chikou Span')
plt.plot(df['Close'], label='Close Price')  # Add close price for reference
plt.legend()
plt.show()
````

# **MA crossover detection**

````python
df = pd.read_csv('AAPL.csv')

# Assuming your CSV has a 'Date' column and 'Close' price data
df['Date'] = pd.to_datetime(df['Date'])  # Convert 'Date' to datetime if necessary
df.set_index('Date', inplace=True)

# Calculate Moving Averages
df['SMA_50'] = df['Close'].rolling(window=50).mean()
df['SMA_200'] = df['Close'].rolling(window=200).mean()

# Method 1: Calculation and Comparison
df['Crossover'] = np.where(df['SMA_50'] > df['SMA_200'], 1, 0)  # 1 for bullish crossover
df['Crossover'] = np.where(df['SMA_50'] < df['SMA_200'], -1, df['Crossover'])  # -1 for bearish

# Method 2: Using diff()
df['Diff'] = (df['SMA_50'] - df['SMA_200']).diff()
df['Crossover2'] = np.where(df['Diff'] > 0, 1, np.where(df['Diff'] < 0, -1, 0))

# Plot the results
plt.figure(figsize=(20, 6))
plt.plot(df['Close'], label='Close')
plt.plot(df['SMA_50'], label='SMA 50')
plt.plot(df['SMA_200'], label='SMA 200')

# Mark crossovers (using either 'Crossover' or 'Crossover2' column)
plt.scatter(df.index[df['Crossover'] == 1], df.loc[df['Crossover'] == 1, 'Close'], marker='^', color='green')
plt.scatter(df.index[df['Crossover'] == -1], df.loc[df['Crossover'] == -1, 'Close'], marker='v', color='red')

plt.legend()
plt.show()
````

# **Guppy Multiple Moving Average**

````python
import pandas as pd
import matplotlib.pyplot as plt

def calculate_gmma(df, short_periods, long_periods):
    """Calculates the Guppy Multiple Moving Average (GMMA).

    Args:
        df (pandas.DataFrame): Dataframe containing stock data (must have a 'Close' column).
        short_periods (list): List of periods for the short-term EMAs.
        long_periods (list): List of periods for the long-term EMAs.

    Returns:
        pandas.DataFrame: The original DataFrame with new GMMA columns added.
    """

    for period in short_periods:
        df['EMA_' + str(period)] = df['Close'].ewm(span=period, adjust=False).mean()

    for period in long_periods:
        df['EMA_' + str(period)] = df['Close'].ewm(span=period, adjust=False).mean()

    return df

def plot_gmma(df, short_periods, long_periods):
    """Plots the GMMA along with the price."""

    plt.figure(figsize=(20, 6))

    # Plot price
    plt.plot(df['Close'], label='Close')

    # Plot short-term EMAs
    for period in short_periods:
        plt.plot(df['EMA_' + str(period)], label=f'EMA {period} (Short)')

    # Plot long-term EMAs
    for period in long_periods:
        plt.plot(df['EMA_' + str(period)], label=f'EMA {period} (Long)')

    plt.legend()
    plt.show()

# Example Usage
df = pd.read_csv('AAPL.csv')
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

short_periods = [3, 5, 8, 10, 12, 15]
long_periods = [30, 35, 40, 45, 50, 60]

df = calculate_gmma(df, short_periods, long_periods)
plot_gmma(df, short_periods, long_periods)
````

# **Chop Zone**

````python
def chop_zone(df, upper_threshold=61.8, lower_threshold=38.2, chop_color='turquoise', trend_color='gray'):
    """Calculates the Chop Zone indicator and adds visualization columns.

    Args:
        df (pandas.DataFrame): DataFrame containing stock data (must have 'High', 'Low', 'Close' columns).
        upper_threshold (float): Upper threshold for the Choppiness Index.
        lower_threshold (float): Lower threshold for the Choppiness Index.
        chop_color (str): Color for the chop zone.
        trend_color (str): Color for the trending zone.
    """

    def true_range(df):
        tr1 = df['High'] - df['Low']
        tr2 = abs(df['High'] - df['Close'].shift())
        tr3 = abs(df['Low'] - df['Close'].shift())
        return pd.DataFrame({'tr1': tr1, 'tr2': tr2, 'tr3': tr3}).max(axis=1)

    def choppiness_index(df, n=14):
        atr = true_range(df).rolling(n).sum()
        high_n = df['High'].rolling(n).max()
        low_n = df['Low'].rolling(n).min()
        ci = 100 * np.log10((atr.sum() / (high_n - low_n))) / np.log10(n)
        return ci

    df['ChopIndex'] = choppiness_index(df)
    df['Zone'] = trend_color  # Initialize with trend color

    df.loc[(df['ChopIndex'] > upper_threshold), 'Zone'] = chop_color
    df.loc[(df['ChopIndex'] < lower_threshold), 'Zone'] = chop_color

# Example Usage
df = pd.read_csv('AAPL.csv')
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

chop_zone(df)

# Plotting (Simplified)
plt.figure(figsize=(19, 6))
plt.plot(df['Close'])

# Fill the zones with appropriate colors
for i in range(len(df)):
    plt.fill_between(df.index[i:i+2], df['Close'][i:i+2], 0, where=(df['Zone'][i:i+2] == 'turquoise'), color='turquoise', alpha=0.4)
    plt.fill_between(df.index[i:i+2], df['Close'][i:i+2], 0, where=(df['Zone'][i:i+2] == 'gray'), color='gray', alpha=0.2)

plt.show()
````

# **Chop Zone**

````python
def donchian_channels(df, period=20):
    """Calculates the Donchian Channels.

    Args:
        df (pandas.DataFrame): DataFrame with stock data (must have 'High', 'Low' columns).
        period (int): Lookback period for calculating the channels.

    Returns:
        pandas.DataFrame: The original DataFrame with new Donchian Channel columns.
    """

    df['Upper'] = df['High'].rolling(period).max()
    df['Middle'] = (df['High'].rolling(period).max() + df['Low'].rolling(period).min()) / 2
    df['Lower'] = df['Low'].rolling(period).min()

    return df

# Example Usage
df = pd.read_csv('AAPL.csv')
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

df = donchian_channels(df)

# Plotting
plt.figure(figsize=(20, 6))
plt.plot(df['Close'], label='Close')
plt.plot(df['Upper'], label='Upper', linestyle='--')
plt.plot(df['Middle'], label='Middle', linestyle='--')
plt.plot(df['Lower'], label='Lower', linestyle='--')
plt.fill_between(df.index, df['Upper'], df['Lower'], alpha=0.2)
plt.legend()
plt.show()
````

````python
# ------------------ Guppy Multiple Moving Average (GMMA) ------------------

def calculate_gmma(df, short_periods, long_periods):
    """Calculates the Guppy Multiple Moving Average (GMMA).

    Args:
        df (pandas.DataFrame): Dataframe containing stock data (must have a 'Close' column).
        short_periods (list): List of periods for the short-term EMAs.
        long_periods (list): List of periods for the long-term EMAs.

    Returns:
        pandas.DataFrame: The original DataFrame with new GMMA columns added.
    """

    for period in short_periods:
        df['EMA_' + str(period)] = df['Close'].ewm(span=period, adjust=False).mean()

    for period in long_periods:
        df['EMA_' + str(period)] = df['Close'].ewm(span=period, adjust=False).mean()

    return df

def plot_gmma(df, short_periods, long_periods):
    """Plots the GMMA along with the price."""

    plt.figure(figsize=(12, 6))

    # Plot price
    plt.plot(df['Close'], label='Close')

    # Plot short-term EMAs
    for period in short_periods:
        plt.plot(df['EMA_' + str(period)], label=f'EMA {period} (Short)')

    # Plot long-term EMAs
    for period in long_periods:
        plt.plot(df['EMA_' + str(period)], label=f'EMA {period} (Long)')

    plt.legend()
    plt.show()

# ------------------------------ Chop Zone ---------------------------------

def chop_zone(df, upper_threshold=61.8, lower_threshold=38.2, chop_color='turquoise', trend_color='gray'):
    """Calculates the Chop Zone indicator and adds visualization columns.

    Args:
        df (pandas.DataFrame): DataFrame containing stock data (must have 'High', 'Low', 'Close' columns).
        upper_threshold (float): Upper threshold for the Choppiness Index.
        lower_threshold (float): Lower threshold for the Choppiness Index.
        chop_color (str): Color for the chop zone.
        trend_color (str): Color for the trending zone.
    """

    def true_range(df):
        tr1 = df['High'] - df['Low']
        tr2 = abs(df['High'] - df['Close'].shift())
        tr3 = abs(df['Low'] - df['Close'].shift())
        return pd.DataFrame({'tr1': tr1, 'tr2': tr2, 'tr3': tr3}).max(axis=1)

    def choppiness_index(df, n=14):
        atr = true_range(df).rolling(n).sum()
        high_n = df['High'].rolling(n).max()
        low_n = df['Low'].rolling(n).min()
        ci = 100 * np.log10((atr.sum() / (high_n - low_n))) / np.log10(n)
        return ci

    df['ChopIndex'] = choppiness_index(df)
    df['Zone'] = trend_color  # Initialize with trend color

    df.loc[(df['ChopIndex'] > upper_threshold), 'Zone'] = chop_color
    df.loc[(df['ChopIndex'] < lower_threshold), 'Zone'] = chop_color

# -------------------------- Donchian Channels -----------------------------

def donchian_channels(df, period=20):

    """Calculates the Donchian Channels.

    Args:
        df (pandas.DataFrame): DataFrame with stock data (must have 'High', 'Low' columns).
        period (int): Lookback period for calculating the channels.

    Returns:
        pandas.DataFrame: The original DataFrame with new Donchian Channel columns.
    """

    df['Upper'] = df['High'].rolling(period).max()
    df['Middle'] = (df['High'].rolling(period).max() + df['Low'].rolling(period).min()) / 2
    df['Lower'] = df['Low'].rolling(period).min()

    return df

# ---------------------------- Main Code -----------------------------------

# Load your data
df = pd.read_csv('AAPL.csv')
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Calculate Indicators
df['SMA_50'] = df['Close'].rolling(window=50).mean()
df['SMA_200'] = df['Close'].rolling(window=200).mean()
df['Crossover'] = np.where(df['SMA_50'] > df['SMA_200'], 1, 0)
df['Crossover'] = np.where(df['SMA_50'] < df['SMA_200'], -1, df['Crossover'])

short_periods = [3, 5, 8, 10, 12, 15]
long_periods = [30, 35, 40, 45, 50, 60]
df = calculate_gmma(df, short_periods, long_periods)
chop_zone(df)
df = donchian_channels(df)

# Plotting
plt.figure(figsize=(20, 9))

# Main Price Plot
plt.subplot(2, 1, 1)
plt.plot(df['Close'], label='Close')
plt.plot(df['SMA_50'], label='SMA 50')
plt.plot(df['SMA_200'], label='SMA 200')
plt.scatter(df.index[df['Crossover'] == 1], df.loc[df['Crossover'] == 1, 'Close'], marker='^', color='green')
plt.scatter(df.index[df['Crossover'] == -1], df.loc[df['Crossover'] == -1, 'Close'], marker='v', color='red')
plot_gmma(df, short_periods, long_periods)
plt.legend(loc='upper left')

# Chop Zone and Donchian Channels
plt.subplot(2, 1, 2)
plt.plot(df['Close'])
plt.plot(df['Upper'], label='Upper', linestyle='--')
plt.plot(df['Middle'], label='Middle', linestyle='--')
plt.plot(df['Lower'], label='Lower', linestyle='--')
plt.fill_between(df.index, df['Upper'], df['Lower'], alpha=0.2)

for i in range(len(df)):
    plt.fill_between(df.index[i:i+2], df['Close'][i:i+2], 0, where=(df['Zone'][i:i+2] == 'turquoise'), color='turquoise', alpha=0.4)
    plt.fill_between(df.index[i:i+2], df['Close'][i:i+2], 0, where=(df['Zone'][i:i+2] == 'gray'), color='gray', alpha=0.2)

plt.legend(loc='upper left')
plt.show()
````
